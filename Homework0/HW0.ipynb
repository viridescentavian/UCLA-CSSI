```python
import matplotlib.pyplot as plt
import numpy as np

# matplotlib practice using different functions with colors and labels

# setting x-axis
x = np.linspace(0, 2, 100)

# y=x
linY = x
plt.plot(x, linY, '#4287f5', label='linear')

# y=x^2
quadY = x**2
plt.plot(x, quadY, '#8df7c2', label='quadratic')

# y=x^3
cubY = x**3
plt.plot(x, cubY, '#55d136', label='cubic')

# y=sin(x)
sinY = np.sin(x)
plt.plot(x, sinY, '#171fff', label='sine')

# y=cos(x)
cosY = np.cos(x)
plt.plot(x, cosY, '#8620d4', label='cosine')

# graph labels
plt.legend(loc='upper right')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```
